# Episode 1: The Old Grist Mill




```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('../data/cleaned_over_the_garden_wall_dialogue.csv')

# Filter for just Episode 1
episode_3_df = df[df['episode'] == 'Episode 3: Schooltown Follies'].copy()
```

```python
# Print all dialogue from episode 2
print("\nEpisode 3 Dialogue:")
print("-" * 50)
for _, row in episode_3_df.iterrows():
    line = row['line']
    # Split line if longer than 60 chars
    if len(line) > 60:
        words = line.split()
        new_line = ''
        current_line = ''
        
        for word in words:
            if len(current_line + ' ' + word) > 60:
                new_line += current_line + '\n'
                current_line = word
            else:
                current_line += ' ' + word if current_line else word
                
        new_line += current_line
        print(f"{row['char_name']}: {new_line}")
    else:
        print(f"{row['char_name']}: {line}")
    print("-" * 50)
```
